### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".


# Show the top 5 results.
athletic_sales_2020_df = pd.read_csv('Resources/athletic_sales_2020.csv')
athletic_sales_2021_df = pd.read_csv('Resources/athletic_sales_2021.csv')
result = pd.concat([athletic_sales_2020_df, athletic_sales_2021_df])

print("Available columns:")
print(result.columns)
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.


# Rename the "units_sold" column to "Total_Products_Sold"


# Show the top 5 results.
# Combine the DataFrames
frames = [athletic_sales_2020_df, athletic_sales_2021_df]
result = pd.concat(frames)

# Reset the index
result.reset_index(drop=True, inplace=True)

# Group by region, state, and city, sum the units sold
products_sold = result.groupby(['region', 'state', 'city'])['units_sold'].sum().reset_index()

# Rename the "units_sold" column to "Total_Products_Sold"
products_sold = products_sold.rename(columns={'units_sold': 'Total_Products_Sold'})

# Sort the results in descending order of Total_Products_Sold
products_sold = products_sold.sort_values('Total_Products_Sold', ascending=False)

# Show the top 5 results
print("Top 5 regions/states/cities by total products sold:")
print(products_sold.head())
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.


# Optional: Rename the "total_sales" column to "Total Sales"


# Show the table.
# Combine the two DataFrames if not already done
combined_df = pd.concat([athletic_sales_2020_df, athletic_sales_2021_df], ignore_index=True)

# Create the pivot table
pivot_table = pd.pivot_table(combined_df, 
                             values='total_sales', 
                             index='invoice_date', 
                             aggfunc='sum')

# Rename the "total_sales" column to "Total Sales"
pivot_table = pivot_table.rename(columns={'total_sales': 'Total Sales'})

# Show the table
pivot_table
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.


# Sort the resampled pivot table in descending order on "Total Sales".
# Combine the two DataFrames if not already done
combined_df = pd.concat([athletic_sales_2020_df, athletic_sales_2021_df], ignore_index=True)

# Create the pivot table
pivot_table = pd.pivot_table(combined_df, 
                             values='total_sales', 
                             index='invoice_date', 
                             aggfunc='sum')

# Rename the "total_sales" column to "Total Sales"
pivot_table = pivot_table.rename(columns={'total_sales': 'Total Sales'})

# Ensure the index is in datetime format
pivot_table.index = pd.to_datetime(pivot_table.index)

# Resample into daily bins and sum the total sales
daily_sales = pivot_table.resample('D').sum()

# Sort the resampled pivot table in descending order on "Total Sales"
daily_sales_sorted = daily_sales.sort_values('Total Sales', ascending=False)

# Show the table
daily_sales_sorted
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.


# Sort the resampled pivot table in descending order on "Total Sales".
# Combine the two DataFrames if not already done
combined_df = pd.concat([athletic_sales_2020_df, athletic_sales_2021_df], ignore_index=True)

# Create the pivot table
pivot_table = pd.pivot_table(combined_df, 
                             values='total_sales', 
                             index='invoice_date', 
                             aggfunc='sum')

# Rename the "total_sales" column to "Total Sales"
pivot_table = pivot_table.rename(columns={'total_sales': 'Total Sales'})

# Ensure the index is in datetime format
pivot_table.index = pd.to_datetime(pivot_table.index)

# Resample into weekly bins and sum the total sales
weekly_sales = pivot_table.resample('W').sum()

# Sort the resampled pivot table in descending order on "Total Sales"
weekly_sales_sorted = weekly_sales.sort_values('Total Sales', ascending=False)

# Show the table
weekly_sales_sorted
```
